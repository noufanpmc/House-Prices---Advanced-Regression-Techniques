```python
# Identify the list if numerical variables #

features_numerical = [i for i in df_train.columns if df_train[i].dtype!='O']

features_numerical
```

```python
# Categorical variables #

categorical_features = [i for i in df_train.columns if df_train[i].dtypes=='O']

categorical_features
```

```python
# Same way we can identify the datatime featurws, we can help us to identify the no. of days  / no. of years between
# the house was built and when it was sold #

year_feature =[i for i in df_train.columns if 'Yr' in i or 'Year' in i]
year_feature
```

```python
# We will droop the ID column and the variables : Alley, fence, poolQc, fence and Miscfeature where the missjng values are greater than 50%
# in both the train and test data
# Train data
df_n = df_train.drop(['Id','Alley','PoolQC','Fence','MiscFeature'],axis=1)
df_n.shape
```

```python
# test data
df1_n = df_test.drop(['Id','Alley','PoolQC','Fence','MiscFeature'],axis=1)
df1_n.shape
```

```python
# Train Data # Imputing the categorical variables with mode values 

for i in df_n.columns:
    if df_n[i].dtype=='O':
        df_n[i]= df_n[i].fillna(df_n[i].mode()[0])
```

```python
# test Data # Imouting the categorical variables with mode values 

for i in df1_n.columns:
    if df1_n[i].dtype=='O':
        df1_n[i]= df1_n[i].fillna(df1_n[i].mode()[0])
```

```python
# Train data, for the numerical variables where we will replace missing 
# values with median instead of mean since, we know that there are outliers in the data #

for i in df_n.columns:
    if df_n[i].dtype=='float64' or df_n[i].dtype=='int64':
        df_n[i]= df_n[i].fillna(df_n[i].median())
```

```python
#test data
for i in df1_n.columns:
    if df1_n[i].dtype=='float64' or df1_n[i].dtype=='int64':
        df1_n[i]= df1_n[i].fillna(df1_n[i].median())
```

```python
print(df1_n.isnull().sum().sum())
print(df_n.isnull().sum().sum())
```

```python
# Temporal variables #
# Train data #
for i in year_feature:
    if i!='YrSold':
    # this will give the no. of days #
        print("for the feature",i)
        df_n[i] = df_n['YrSold'] - df_n[i]
df_n.head()
```

```python
df_n[year_feature].head()
```

```python
# Test data #
for i in year_feature:
    if i!='YrSold':
    # this will give the no. of days #
        df1_n[i] = df1_n['YrSold'] - df1_n[i]
df1_n.head()
```

```python
df1_n[year_feature].head()
```

```python
y_train = df_n['SalePrice']
x_train = df_n.drop(['SalePrice'],axis=1)
x_train.shape
```

```python
# Lets combine rhe data from train and test #

df_new = pd.concat([x_train,df1_n],axis=0)
df_new.tail()
```

```python
df_new.shape
```

```python
df_cat = pd.get_dummies(df_new,drop_first=True)
df_cat.head()
```

```python
# Splitting the total daata set back into train and test back


x_train = df_cat.iloc[:1460,:]

x_train.shape
```

```python
x_test =  df_cat.iloc[1460:,:]

x_test.shape
```

```python
from xgboost import XGBRegressor

reg = xgboost.XGBRegressor()
reg.fit(x_train,y_train)
```

```python
# Predicting the test data set #

y_pred = reg.predict(x_test)
y_pred
```

```python
# Converting this array into data frame #

y_pr = pd.DataFrame(y_pred)
y_pr.head()
```

```python
# Hyoer parameters #
n_estimators = [100,500,900,1000,1200,1500]
max_depth=[2,3,5,10,15]
learning_rate=[0.05,0.1,0.15,0.2,0.3]
min_child_weight=[1,2,3,4]
base_score = [0.25,0.5,0.75,1]
booster =['gbtree','gblinear']

hyp_grid = {'n_estimators':n_estimators,'max_depth':max_depth,'learning_rate':learning_rate,'min_child_weight':min_child_weight,
            'booster':booster,'base_score':base_score}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
regressor =  xgboost.XGBRegressor()
```

```python
# Random Search CV for this model #

random_cv = RandomizedSearchCV(estimator=regressor,param_distributions=hyp_grid,cv=5,n_iter=50,
                               scoring='neg_mean_absolute_error',n_jobs=4,verbose=5,return_train_score=True,
                               random_state=1)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```
